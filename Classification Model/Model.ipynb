```python
# Encoding the species column

from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
```

```python
df['species'] = le.fit_transform(df['species'])
```

# Building the Models

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,recall_score,precision_score,confusion_matrix,roc_curve
```

```python
x = df.drop(columns = 'species')
y = df['species']
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

### Logistic Regression

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print(f"Accuracy: {accuracy_score(y_test,lr.predict(x_test))}")
print(f"Precision: {precision_score(y_test,lr.predict(x_test),average='weighted')}")
print(f"Recall: {recall_score(y_test,lr.predict(x_test),average='weighted')}")
```

```python
sns.heatmap(confusion_matrix(y_test,lr.predict(x_test)),annot =True)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
